```python
import pandas as pd

# Paso 1: Generar un DataFrame con los datos de los cuatro archivos.
emisiones_2016 = pd.read_csv('emisiones-2016.csv', sep=';')
emisiones_2017 = pd.read_csv('emisiones-2017.csv', sep=';')
emisiones_2018 = pd.read_csv('emisiones-2018.csv', sep=';')
emisiones_2019 = pd.read_csv('emisiones-2019.csv', sep=';')

# Combinar los DataFrames en uno solo
df = pd.concat([emisiones_2016, emisiones_2017, emisiones_2018, emisiones_2019])
```

```python
# Seleccionamos las columnas deseadas y las almacenamos en una lista.
# Luego, filtramos el DataFrame para incluir solo estas columnas.
columnas_filtradas = ['ESTACION', 'MAGNITUD', 'ANO', 'MES'] + [col for col in df.columns if col.startswith('D')]
df = df[columnas_filtradas]
```

```python
# Paso 3: Reestructurar el DataFrame
# Utilizamos df.melt() para reestructurar el DataFrame de formato ancho a formato largo, manteniendo las columnas ESTACION, MAGNITUD, AÑO y MES.
df = df.melt(id_vars=['ESTACION', 'MAGNITUD', 'ANO', 'MES'], var_name='DIA', value_name='EMISION')
```

```python
# Paso 4: Añadir una columna con la fecha
# Utilizamos pd.to_datetime() para convertir las columnas de año, mes y día en objetos de fecha y hora.
# Concatenamos las columnas de año, mes y día, asegurándonos de que estén formateadas correctamente.
# Usamos errors='coerce' para manejar posibles errores de conversión, convirtiendo fechas inválidas en NaT (Not a Time).
df['FECHA'] = pd.to_datetime(df['ANO'].astype(str) + df['MES'].astype(str).str.zfill(2) + df['DIA'].str[1:].str.zfill(2), format='%Y%m%d', errors='coerce')
```

```python
# Paso 5: Eliminar las filas con fechas no válidas
import numpy as np

# Eliminar las filas con fechas no válidas utilizando la función isnat de numpy
df = df[~np.isnat(df['FECHA'])]

# Ordenar el DataFrame por estaciones contaminantes y fecha
df = df.sort_values(by=['ESTACION', 'MAGNITUD', 'FECHA'])

# Mostrar el DataFrame ordenado
print(df)
```

```python
# Paso 6: Mostrar las estaciones y contaminantes disponibles
print("Estaciones disponibles:")
print(df['ESTACION'].unique())
print("\nContaminantes disponibles:")
print(df['MAGNITUD'].unique())
print("\n")
```

```python
# Paso 7: Función para obtener las emisiones por estación, contaminante y rango de fechas
def obtener_emisiones(estacion, contaminante, fecha_inicio, fecha_fin):
    return df[(df['ESTACION'] == estacion) & (df['MAGNITUD'] == contaminante) &
              (df['FECHA'] >= fecha_inicio) & (df['FECHA'] <= fecha_fin)]['EMISION']
```

```python
# Paso 8: Resumen descriptivo para cada contaminante
resumen_contaminantes = df.groupby('MAGNITUD')['EMISION'].describe()
print("Resumen descriptivo para cada contaminante:")
print(resumen_contaminantes)
print("\n")
```

```python
# Paso 9: Resumen descriptivo para cada contaminante por estación
resumen_contaminantes_por_estacion = df.groupby(['ESTACION', 'MAGNITUD'])['EMISION'].describe()
print("Resumen descriptivo para cada contaminante por estación:")
print(resumen_contaminantes_por_estacion)
print("\n")
```

```python
# Paso 10: Función para obtener resumen descriptivo de emisiones por contaminante y estación
def resumen_descriptivo_emisiones(estacion, contaminante):
    return df[(df['ESTACION'] == estacion) & (df['MAGNITUD'] == contaminante)]['EMISION'].describe()
```

```python
# Paso 11: Función para obtener las emisiones medias mensuales de un contaminante y un año dados para todas las estaciones
def emisiones_medias_mensuales(contaminante, anio):
    return df[(df['MAGNITUD'] == contaminante) & (df['ANO'] == anio)].groupby('MES')['EMISION'].mean()
```

```python
# Paso 12: Función para obtener medias mensuales de los distintos tipos de contaminantes para una estación dada
def medias_mensuales_por_contaminante(estacion):
    return df[df['ESTACION'] == estacion].groupby(['MAGNITUD', 'MES'])['EMISION'].mean()
```
